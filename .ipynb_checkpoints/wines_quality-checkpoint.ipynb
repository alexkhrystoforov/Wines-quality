```python
train['A'] = train['A'].fillna(train['A'].mode()[0])
train['H'] = train['H'].fillna(train['H'].mode()[0])
```

explore target deeper

```python
sns.countplot(x='Target', data=train)
plt.show()
```

We deal with imbalanced classes.Our target mostly scattered from to 3 to 5. 4 is the most common (near 2500 times).
we 

```python
corrmat = train.corr()
k = 18
cols = corrmat.nlargest(k, 'Target')['Target'].index
plt.subplots(figsize=(20, 10))
coef = np.corrcoef(train[cols].values.T)
sns.set(font_scale=1.25)
mask = np.triu(np.ones_like(corrmat, dtype=np.bool))
sns.heatmap(coef, mask=mask, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

We don't have any strong correlation with our Target, only Alcohol has some. But we see strong correlation berween D and Di.

```python
plt.figure(figsize = (10,6))
sns.countplot(x = 'Type', data = train)
plt.show()
```

```python
sns.catplot(x = "Target", y = "Alcohol", data = train, height = 5, aspect = 3)
plt.title("Alcohol percent in each level of wine's quality")
plt.show()
```

drop 1 outlier

```python
train = train.drop(train[train['Alcohol'] > 10.5].index)
```

```python
# train['Type'] = pd.get_dummies(train['Type'])
```

```python
train_y = train['Target']
del train['Target']
```

## MODELING

```python
X_train, X_test, y_train, y_test = train_test_split(train, train_y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.tree import DecisionTreeClassifier
# from sklearn.linear_model import LogisticRegression
# from sklearn.base import BaseEstimator

# RF = RandomForestClassifier()
# KNN = KNeighborsClassifier()
# TREE = DecisionTreeClassifier()
# LOG = LogisticRegression()
```

```python
RF = RandomForestClassifier()

RF.fit(X_train, y_train)

X_test_y = RF.predict(X_test)

sklearn.metrics.accuracy_score(y_test, X_test_y)
```
